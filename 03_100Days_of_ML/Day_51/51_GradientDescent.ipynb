```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('placement.csv')
```

```python
x=df.iloc[:,0].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

### Prediction using sklearn LinearRegression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train.reshape((-1,1)),y_train)
print(lr.coef_)
print(lr.intercept_)
```

```python
y_pred=lr.predict(x_test.reshape((-1,1)))
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

### Prediction using myGradientDescent

```python
class myGradientDescent :
    def   __init__(self,learning_rate,epochs):
        self.b = 0
        self.m = 0
        self.epoch=epochs
        self.lr=learning_rate
        
    def fit(self,X,Y):
        X = X.ravel()
        Y = Y.ravel()
        n = len(X)
        for i in range(self.epoch):
            slope_m = (-2/n) * np.sum((Y - self.m * X - self.b)*X)
            slope_b = (-2/n) * np.sum(Y - self.m * X - self.b)
            
            self.m=self.m-(self.lr*slope_m)
            self.b=self.b- (self.lr*slope_b)
       
        print(self.m , self.b)


    def predict(self,x_test):
        return self.m*x_test+self.b
```

```python
myGD=myGradientDescent(0.01,10000)
```

```python
myGD.fit(x_train.reshape((-1,1)),y_train)
```

```python
y_pred2=myGD.predict(x_test.reshape((-1,1)))
```

```python
r2_score(y_test,y_pred2)
```
