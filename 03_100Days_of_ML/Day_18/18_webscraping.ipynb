```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 6.3; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/80.0.3987.162 Safari/537.36"
    )
}

pages=[]

for i in range(1, 8):
    url = f"https://www.ambitionbox.com/list-of-companies?designation=data-scientist&sortBy=topPaying&experience=0&page={i}"
    response = requests.get(url, headers=headers)
    pages.append(response.text)

```

## Finding out the Data Science Job Oppurtunities for Freshers

```python
name=[]
rating=[]
avg_salary=[]
location=[]
```

```python
for webpage in pages:
    soup=BeautifulSoup(webpage,'lxml')
    company=soup.find_all('div',class_='companyCardWrapper')
    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('div',class_='rating_text').text.strip())
        avg_salary.append(i.find('div',class_='detailedRange').text.split()[1])
        location.append(i.find('span',class_='companyCardWrapper__interLinking').text.strip().partition('|')[2].partition('/')[0].strip().partition('+')[0])

```

```python
companyData={
    'name': name,
    'rating': rating,
    'avg_salary (L/year)' : avg_salary,
    'location':location
}

df=pd.DataFrame(companyData)
```

```python
df.to_csv("DataScience_Jobs_Freshers")
```
