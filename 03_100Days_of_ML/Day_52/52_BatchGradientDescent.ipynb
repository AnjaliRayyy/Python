```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('advertising.csv')
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

### Using sklearn LinearRegression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
print(lr.intercept_)
print(lr.coef_)
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred=lr.predict(x_test)
r2_score(y_test,y_pred)
```

### Using Batch Gradient Descent coded from scratch

```python
class GDRegressor:
    def __init__(self,learning_rate=0.001,epochs=100):
        self.intercept_=None
        self.coef_= None
        self.lr = learning_rate
        self.epoch = epochs

    def fit(self,X,Y):
        self.intercept_=120
        self.coef_=np.ones(X.shape[1])

        for i in range(self.epoch):
            # updating intercept
            y_hat=self.intercept_ + np.dot(X,self.coef_)
            intercept_der=-2 * np.mean(Y-y_hat)
            self.intercept_ = self.intercept_ - (self.lr * intercept_der)

            # updating coefficients
            coef_der = -(2/X.shape[0]) * (np.dot(Y-y_hat,X))
            self.coef_ =self.coef_ - (self.lr * coef_der)

    def predict(self,X):
        return np.dot(X,self.coef_)+self.intercept_
```

```python
myGD=GDRegressor(learning_rate=0.01,epochs=10000)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
myGD.fit(x_train,y_train)
```

```python
myGD.intercept_
```

```python
myGD.coef_
```

```python
y_pred2=myGD.predict(x_test)
```

```python
r2_score(y_test,y_pred2)
```
