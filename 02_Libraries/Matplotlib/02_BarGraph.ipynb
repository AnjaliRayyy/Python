```python
import matplotlib.pyplot as plt
import numpy as np
```

### Vertical plot

```python
labels=list("ABCDE")
values=[2,10,5,4,10]

bars=plt.bar(labels,values,color="b",width=0.5)

patterns=['/','x','/','x','/']
for bar,pattern in zip(bars,patterns):
    bar.set_hatch(pattern)


plt.title("Bar Graph")
plt.xlabel("X-label")
plt.ylabel("Y-label")

plt.show()
```

### Horizontal Plot

```python
labels=list("ABCDE")
values=[2,10,5,4,10]

bars=plt.barh(labels,values,color="hotpink",height=0.5)

patterns=['/','x','/','x','/']
for bar,pattern in zip(bars,patterns):
    bar.set_hatch(pattern)

plt.title("Bar Graph")
plt.xlabel("X-label")
plt.ylabel("Y-label")

plt.show()
```
