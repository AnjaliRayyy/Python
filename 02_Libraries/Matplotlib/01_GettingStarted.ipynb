### Loading Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=[0,1,2,3,4]
y=[0,1,2,3,4]

#Resize your graph
plt.figure(figsize=(5,3),dpi=100)

#plt.plot(x,y,label='value',color='red',linewidth=2,marker='.',markersize=10)
# Shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y,'bo:',mec='blue',mfc='b',label='2x')

# Ploting another line
x2=np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,'ro-',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'ro--',label='x^2')

plt.title('First Graph',fontdict={'fontsize':20})
plt.xlabel("X-Label")
plt.ylabel("Y-Label")

plt.xticks([0,1,2,3,4])
plt.yticks(np.arange(0,17,2))

plt.legend()
plt.savefig('mygraph.png',dpi=300)
plt.show()
```
