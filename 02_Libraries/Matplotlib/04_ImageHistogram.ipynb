# Histogram Equalization

### Importing necessary libraries

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

### Processing the image

```python
img=cv.imread('./sample_img.jpg',cv.IMREAD_GRAYSCALE)
plt.imshow(img,cmap='gray')
plt.axis('off')
plt.show()
```

### Displaying image histogram

```python
hist=cv.calcHist([img],[0],None,[256],[0,256])
plt.figure()
plt.plot(hist)
```

```python
clrimg=cv.imread('./sample_img.jpg')
imgRGB=cv.cvtColor(clrimg,cv.COLOR_BGR2RGB)
plt.figure()
plt.imshow(imgRGB)
plt.axis('off')
plt.show()
```

```python
colors=['b','g','r']
plt.figure()
for i in range(len(colors)):
    hist=cv.calcHist([imgRGB],[i],None,[256],[0,256])
    plt.plot(hist,colors[i])
plt.xlabel('pixel intensity')
plt.ylabel('No of pixels')
plt.show()
```

### Equalizing the image

```python
equ=cv.equalizeHist(img)
res=np.hstack((img,equ))
plt.imshow(res,cmap='gray')
plt.title("Original Image vs Equalized Image")
plt.axis('off')
plt.show()
```

### Historgram of equalized image

```python
histequ=cv.calcHist([equ],[0],None,[256],[0,256])
plt.plot(histequ)
plt.title("Histogram of Equalized Image")
plt.xlabel('Pixel intensity')
plt.ylabel('No. of pixels')
plt.show()
```

### Side by side comparision

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.xlabel('Pixel intensity')
plt.ylabel("No. of pixels")
plt.title("Histogram of original image")
plt.plot(hist)

plt.subplot(1,2,2)
plt.xlabel('Pixel intensity')
plt.ylabel("No. of pixels")
plt.title("Histogram of equalized image")
plt.plot(histequ)

plt.show()
```
