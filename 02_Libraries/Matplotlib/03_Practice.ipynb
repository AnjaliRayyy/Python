## Loading Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Real World Examples #1

### Loading the data

```python
gas=pd.read_csv('gas_prices.csv')
gas
```

### A simple plot

```python
plt.plot((gas.Year),gas.Canada)
plt.show()
```

### Plotting for multiple countries

```python
plt.figure(figsize=(8,5), dpi=100)
plt.title('Gas Prices over Time (in USD)')
plt.xlabel('Year')
plt.ylabel('US Dollars')
for cnt in list(gas.columns[1::2]):
    plt.plot(gas.Year,gas[cnt],'.-',label=cnt)

plt.xticks(gas.Year[::3])

plt.legend(fontsize=12,bbox_to_anchor=(1.05,1))
plt.tight_layout()  # Adjust layout to prevent legend from being cut off
plt.savefig('Gas_price_figure.png',dpi=300)
plt.show()
```

## Real World Examples #2

### Loading the data

```python
fifa=pd.read_csv('fifa_data.csv')
fifa.head(5)
```

### Histograms

```python
bins=[40,50,60,70,80,90,100]
plt.hist(fifa.Overall, color="#117CE7",bins=bins)
plt.xticks(bins)

plt.xlabel('Skill Level')
plt.ylabel('Number of Players')
plt.title('Distribution of Player Skills in FIFA 2018')
plt.show()
```

### Pie charts

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count().ID
right=fifa.loc[fifa['Preferred Foot']=='Right'].count().ID
y=[left,right]
labels=['Left','Right']
explode=[0.2,0]
plt.figure(figsize=(7,5))
plt.pie(y,labels=labels,explode=explode,shadow=True)
plt.title('Foot Preference of FIFA Players')
plt.show()


```

```python
fifa.Weight=[int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light=fifa.loc[fifa.Weight < 125].count().ID
light_medium=fifa.loc[(fifa.Weight >=125) & (fifa.Weight < 150)].count().ID
medium=fifa.loc[(fifa.Weight >=150) & (fifa.Weight <175)].count().ID
medium_heavy=fifa.loc[(fifa.Weight >=175) & (fifa.Weight < 200)].count().ID
heavy=fifa.loc[fifa.Weight>=200].count().ID

plt.figure(figsize=(6,6),dpi=100)
plt.style.use('ggplot')
weights=[light,light_medium,medium,medium_heavy,heavy]
labels=['light','light_medium','medium','medium_heavy','heavy']
explode=[0.3,0.3,0,0,0.3]
plt.pie(weights,labels=labels,shadow=True,explode=explode,autopct='%.2f %%')
plt.title("Weight Distribution of FIFA Players (in lbs)")

plt.show()
```

### Box Plot

```python
plt.style.use('default')
plt.figure(figsize=(6,8))
barcelona=fifa.loc[fifa.Club=='FC Barcelona']['Overall']
madrid=fifa.loc[fifa.Club=='Real Madrid']['Overall']
revs=fifa.loc[fifa.Club=='New England Revolution']["Overall"]

labels=['FC Barcelona','Real Madrid','New England Revolution']

boxes=plt.boxplot([barcelona,madrid,revs],tick_labels=labels,patch_artist=True,medianprops={'linewidth':2})

for box in boxes['boxes']:
    box.set(color='#4285f4',linewidth=2)    # edge color
    box.set(facecolor='#e0e0e0')    # changing the face color

plt.title("Professional Soccer Team Comparison")
plt.ylabel('FIFA Overall Rating')
plt.xlabel("Teams")

plt.show()
```
