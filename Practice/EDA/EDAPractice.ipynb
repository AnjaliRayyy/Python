```python
def initials(x):
    if isinstance(x,float):
        return ""
    words=x.split()
    name=[]
    for i in words:
        name.append(i[0])
    return ''.join(name)

x=deliveries['winner']
x_initials=np.vectorize(initials)
team_wins=x_initials(x)
plt.hist(team_wins)
```
