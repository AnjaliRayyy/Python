```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def rename(team):
    result=(x[0] for x in team.split())
    return ''.join(result)
```

```python
df=pd.read_csv('matches.csv',converters={'team1':rename,'team2':rename,'toss_winner':rename,'winner':rename},parse_dates=['date'])
df.drop(columns=['dl_applied','id','venue','city','umpire1','umpire2','umpire3'],inplace=True)
df
```
